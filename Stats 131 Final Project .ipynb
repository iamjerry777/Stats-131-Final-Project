# *Background Information* 

## How baskteball is played
> Modern Basketball game is a team sports in which two teams, five players each, compete against each other shooting basketball into defenders' hoop. Five players on each side fall into five different positions: point guards, shooting guards, small forward, power forward and center. In guard position, Point guard is typically responsible for ball handling, while shooting guard's job is to score the basket. Meanwhile in forward position, small forward typically is a wing player who defends, shoots threes and drives to the basket. A power forward, taller and stronger than a small forward, often does his job inside the three-point line. Last but not the least, a center is usually the tallest player on each team, responsible for rebounding and defending.

> The flow of modern basketball game consists of players dribbling the ball or passing to a teammate. On the offense side, players can create a variety of shots, including lay-up, jump shot or dunk. Typically, a field goal is worth of 2 points unless a player is shooting behind 3-point line. A free throw is worth of 1 point. Players will shoot a pair of free throws when they get fouled and didn't make the shot. But if a player is fouled while still managed to make the shot, he will only shoot once in this case. On the side of defense, a defender can steal or block the ball from an offensive player. A NBA game is normally 48 minutes, equally dividing into 4 quarters, with each quarter of 12 minutes. Teams who score higher in the end of 48 minutes will win. But if there's a tie at the end of 48 minutes, there's going to be an additional 5 minutes overtime in which teams will compete to decide the winner. 

> Typically, a basketball game is longer than 48 minutes because 48 minutes is actually playing time, the clock is stopped while the play is not active. Coaches can call timeout to stop the clock, but the number of timeouts allowed in a game is limited in a gmae. In another case, a foul on the court will stop the clock.

#### Game strategies 

> 1. There are two main defensive strategies: zone defense and man-to-man defense. In a zone defense, each player is assigned to guard a specific area of the court. Zone defenses often allow the defense to double team the ball, a manoeuver known as a trap. In a man-to-man defense, each defensive player guards a specific opponent.(cited from Wikipedia)

> 2. Offensive plays are more varied, normally involving planned passes and movement by players without the ball. A quick movement by an offensive player without the ball to gain an advantageous position is known as a cut. A legal attempt by an offensive player to stop an opponent from guarding a teammate, by standing in the defender's way such that the teammate cuts next to him, is a screen or pick. The two plays are combined in the pick and roll, in which a player sets a pick and then "rolls" away from the pick towards the basket. Screens and cuts are very important in offensive plays; these allow the quick passes and teamwork, which can lead to a successful basket. Teams almost always have several offensive plays planned to ensure their movement is not predictable. On court, the point guard is usually responsible for indicating which play will occur. (cited from Wikipedia)

#### Shooting

> 1. Shooting is the act of attempting to score points by throwing the ball through the basket, methods varying with players and situations.(cited from Wikipedia)

> 2. A player will rest the ball on the fingertips of the dominant hand (the shooting arm) slightly above the head, with the other hand supporting the side of the ball. The ball is usually shot by jumping (though not always) and extending the shooting arm. The shooting arm, fully extended with the wrist fully bent, is held stationary for a moment following the release of the ball, known as a follow-through. Players often try to put a steady backspin on the ball to absorb its impact with the rim. The ideal trajectory of the shot is somewhat controversial, but generally a proper arc is recommended. Players may shoot directly into the basket or may use the backboard to redirect the ball into the basket.(cited from Wikipedia)

> 3. Another common shot is called the lay-up. This shot requires the player to be in motion toward the basket, and to "lay" the ball "up" and into the basket, typically off the backboard (the backboard-free, underhand version is called a finger roll). The most crowd-pleasing and typically highest-percentage accuracy shot is the slam dunk, in which the player jumps very high and throws the ball downward, through the basket while touching it.(cited from Wikipedia)

> 4. A shot that misses both the rim and the backboard completely is referred to as an air ball.

#### Rebounding

> Rebound is to grab the ball or gain possession of the ball after a player on either team misses the shot. A rebound collected by an offensive player whose team misses the previous shot will fall into the category of offensive rebound, a rebound collected by a defender whose opponent misses the previous shot will fall into the category of defensive rebound.

#### Dribbling

> Dribbling is the act of bouncing the ball continuously with one hand, and is a requirement for a player to take steps with the ball. To dribble, a player pushes the ball down towards the ground with the fingertips rather than patting it; this ensures greater control.(Definition cited from Wikipedia)

#### Blocking

> A block is performed when, after a shot is attempted, a defender succeeds in altering the shot by touching the ball.

#### Height

> At the professional level, most male players are above 6 feet 3 inches (1.91 m)

#  *Data Preperation* 
#### 1. The initial dataset has 24691 rows and 53 columns of variables

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
import statsmodels.api as sm
import statsmodels.formula.api as smf

players = pd.read_csv('Seasons_Stats.csv')
players.info()   #Dataset initially has 24691 rows and 53 columns of variables.
```

#### 2. Since the dataset is too big, we filter out the data recorded only after 2010. 

```python
players=players.loc[players.Year>=2010,]      #Statistics after 2010 will be considered
```

#### 3. Drop out blank columns: 'blanl', 'blank2'.

```python
players.drop(["blanl", "blank2"], axis = 1, inplace = True)  #Drop Blank Columns
```

#### 4. Create new variables: 

> PPG: PTS divided by G <br>
  MPG: MP divided by G  <br>
  FGPG: FG divided by G  <br>
  FTPG: FT divided by G  <br>
  ORBPG: ORB divided by G  <br>

```python
players['PPG']=players.PTS/players.G    #Point per game
players['MPG']=players.MP/players.G   #Minute per game
players['FGPG']=players.FG/players.G  #Field goal per game
players['FTPG']=players.FT/players.G    #Free throw per game
players['ORBPG']=players.ORB/players.G  #OFFENSIVE Rebounds per game
```

#### 5. Choose variables we need: Year, Player, Age, GS, PER, TS%, USG%, OWS, WS, FG%

```python
players=players[['Year','Player','Age','GS','PER','TS%','USG%','OWS','WS','FG%','MPG','FGPG','FTPG','ORBPG','PPG']] #We only choose this 15 variables
```

#### 6. Change variable types

```python
#Convert the type of some numerical variables into type integer
players['Year'] = players['Year'].astype(int)   
players['Age'] = players['Age'].astype(int)
players['GS'] = players['GS'].astype(int)  
```

#### 7. Reindex the data.

```python
players.index=np.arange(1,4763,1)
players.head()
```

#  *Data Modeling* 

## Guiding question: what are the significant factors affecting players points per game?

> Initially, we tried different models. Then we decided to use ordinary least square and linear regression model to predict points scored per game, as well as to find the significant factors affecting points scored per game. The reason behind our model selection is that most of our independent variables are numerical, as well as our dependent variable.  <br>

> 1. Fill the missing values with mean value of the columns. <br>
> 2. Extract out the predictors from the original dataset. <br>
> 3. Fit our linear model. <br>
> 4. Use our model to predict the points scored per game.
> 5. Compare our prediction versus actual statistics by using plots.
> 6. Get model score of approximately 0.99.
> 7. Use ordinary least square model to find the significant factors.

```python
#Linear Model
players['target'] = players.PPG               
players=players.fillna(players.mean()) #Fill missing value with the mean
playersmodel = linear_model.LinearRegression() #Fit a linear model
```

```python
X = players.loc[:, 'Age':'ORBPG']  #Get predictors(everything expect PPG)
X=X.fillna(X.mean()) #Fill missing value with the mean
np.where(np.isnan(X))  #Double check if missing values are gone
```

```python
playersmodel.fit(X, players.PPG) #Fit our prediction model
playersmodel.coef_
```

```python
yhat = playersmodel.predict(X)#Predicted PPG 
print(yhat)
```

```python
#Prediction vs actual plot
plt.plot(yhat, players.PPG, 'o')   
plt.xlabel('predicted')
plt.ylabel('actual')
plt.show()
```

```python
#Residual Plot
residuals = players.PPG - yhat
plt.plot(yhat, residuals, 'o')
plt.xlabel('predicted')
plt.ylabel('residual')
plt.show()
```

```python
#Model Score
playersmodel.score(X, players.PPG)
```

```python
model1 = smf.ols('players.PPG ~ X',data=X)
results = model1.fit()
print(results)
print(results.summary())
```

> From the table above, we find that Age and PER are not significant factors affecting points scored per game. This finding justifies our previous assumption about the relationship between age and PPG because as players get older, their points per game have an up and down trend, which is not statistically significant.
