```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
players = pd.read_csv('Seasons_Stats.csv')
```

```python
players=players.loc[players.Year>=2010,]
```

```python
players.drop(["blanl", "blank2"], axis = 1, inplace = True)  #Drop Blank Columns
```

```python
players['PPG']=players.PTS/players.G    #Point per game
```

```python
players['MPG']=players.MP/players.G   #Minute per game
```

```python
players['FGPG']=players.FG/players.G  #Field goal per game
```

```python
players['FTPG']=players.FT/players.G    #Free throw per game
```

```python
players['ORBPG']=players.ORB/players.G  #OFFENSIVE Rebounds per game
```

Variables we choose: Age,GS,PER,TS%,USG%,OWS,WS,FG%,Year,Player
Variables we created:PPG, MPG, FGPG,FTPG,ORBPG,

```python
players=players[['Year','Player','Age','GS','PER','TS%','USG%','OWS','WS','FG%','MPG','FGPG','FTPG','ORBPG','PPG']]
```

```python
players.index=np.arange(1,4763,1)
```

```python
a=players.loc[players['FG%'] >0.5,]
b=a.loc[a['MPG']<5,]
list = np.array(b.index)
players.drop(players.index[list])
#Remove outliers
```

```python
highpoints=players.loc[players.PPG>30,]   #Players with 20 or more points per game
```
